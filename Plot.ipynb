```
# File location and type
file_location = "/FileStore/tables/SMAplot.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
ma_df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

ma_df_spark = ma_df.select('Close','MA1d','MA1w','MA2w','MA1m','MA4m',ma_df.Date.cast('timestamp'))
```

```
ma_df_spark
```

```
ma_df_pandas = ma_df_spark.toPandas()
```

```

# n = 20
# colors = plt.cm.jet(np.linspace(0,1,n))

ax = ma_df_pandas.plot(x='Date',grid=True,figsize = (15,8))
ax.set_xlabel("date")
ax.set_ylabel("USD")
display()
```

```
# File location and type
file_location = "/FileStore/tables/lrexport.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
lr_df = spark.read.format(file_type).option("inferSchema", infer_schema).option("header", first_row_is_header).option("sep", delimiter).load(file_location)

lr_spark = lr_df.select(lr_df.Prediction_1D.alias('lr_1D'),lr_df.Prediction_1W.alias('lr_1W'),lr_df.Prediction_1M.alias('lr_1M'),lr_df.Prediction_4M.alias('lr_4M'),lr_df.Date.cast('timestamp'),'Close')
```

```
# File location and type
file_location = "/FileStore/tables/export_radomforest-485cf.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
rf_df = spark.read.format(file_type).option("inferSchema", infer_schema).option("header", first_row_is_header).option("sep", delimiter).load(file_location)

rf_spark = rf_df.select('Close',rf_df.Prediction_1D.alias('rf_1D'),rf_df.Prediction_1W.alias('rf_1W'),
                        rf_df.Prediction_1M.alias('rf_1M'),rf_df.Prediction_4M.alias('rf_4M'),rf_df.Date.cast('timestamp'))
```

```
# File location and type
file_location = "/FileStore/tables/predict_arima.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
arima_df = spark.read.format(file_type).option("inferSchema", infer_schema).option("header", first_row_is_header).option("sep", delimiter).load(file_location)

arima_spark=arima_df.select('Close',arima_df.day.alias('arima_1D'),arima_df.oneweek.alias('arima_1W'),
                            arima_df.onemonth.alias('arima_1M'),arima_df.fourmonth.alias('arima_4M'),arima_df.Date.cast('timestamp'))
```
