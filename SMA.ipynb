```
import math
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import time

from datetime import date, datetime, time, timedelta
from matplotlib import pyplot as plt
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

%matplotlib inline

# File location and type
file_location = "/FileStore/tables/AAPL__2_-1028f.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
df1 = df.select("*").toPandas()
df1['Close'][165] = df1['Close'][164]
df1['Close'] = pd.to_numeric(df1['Close'])
# Convert Date column to datetime
df1.loc[:, 'Date'] = pd.to_datetime(df1['Date'],format='%Y-%m-%d')


# Get month of each sample
df1['month'] = df1['Date'].dt.month

# Sort by datetime
df1.sort_values(by='Date', inplace=True, ascending=True)
# Plot adjusted close over time
rcParams['figure.figsize'] = 10, 8 # width 10, height 8

ax = df1.plot(x='Date', y='Close', style='b-', grid=True)
ax.set_xlabel("date")
ax.set_ylabel("USD")
display()
```

```
# 1 day moving average model
from sklearn.metrics import mean_squared_error
from math import sqrt
# select most recent data
SMAPE_moving_average = []
# use average closing price from 2 days, 1 week, 2 weeks and 1 month to predict next day
df_list = ['MA1d','MA1w','MA2w','MA1m','MA4m']
c =0 
md1 = df.select("*").toPandas()
md1 = md1[md1 != 'null']
  #restore missing data
MA_df1 = md1[['Date','Close']]
for n_day in (2,5,10,20,80):
  md2 = md1['Close']
  md2 = md2[md2 != 'null']
  md2 = md2.astype('float')
  md2[165] = md2[164]
 
  #md1 = pd.to_numeric(md1['Close'])
  moving_average = []
  i = 0
  avg = 0
  # calulate moving average
  for row in md2:

    if i < n_day:
      moving_average.append(float(md2[i]))
    else:
      for j in range(n_day):
        avg = avg + float((md2[i-j-1]))
      mv = avg/n_day
      avg = 0
      moving_average.append(mv)
    i = i+1
  #rms_moving_average.append(sqrt(mean_squared_error(md2, moving_average)))
  SMAPE_moving_average.append(100/len((md2)) * np.sum(2 * np.abs(np.array(moving_average) - np.array(md2)) / (np.abs(np.array(md2)) + np.abs(np.array(moving_average)))))
  MA1 = moving_average
  MA_df1[df_list[c]] = MA1
  c = c+1
SMAPE_moving_average
```

```
MA_df5= MA_df1
MA_df5['MA1w'] = MA_df1['MA1w'].rolling(window=5).mean()
MA_df5['MA1m'] = MA_df1['MA1m'].rolling(window=26).mean()
MA_df5['MA4m'] = MA_df1['MA4m'].rolling(window=104).mean()
MA_df5 = MA_df5.dropna()
MA_df5['Close'] = MA_df5['Close'].astype(float)
```

```
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))

b=MA_df5.iloc[:,1].values
c=MA_df5.iloc[:,2].values
d=MA_df5.iloc[:,3].values
e=MA_df5.iloc[:,4].values
f=MA_df5.iloc[:,5].values
g=MA_df5.iloc[:,6].values
print(smape(b, c))
print(smape(b, d))
print(smape(b, e))
print(smape(b, f))
print(smape(b, g))
```
